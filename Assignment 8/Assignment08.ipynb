```python
import requests
othelloURL = "https://www.gutenberg.org/cache/epub/2267/pg2267.txt"
r = requests.get(othelloURL,allow_redirects=True)
othelloTxt = r.text
```

```python
specialchars = '\r\n\r\n\r\.,-!:*1234567890<>'
for c in specialchars:
    othelloTxt = othelloTxt.replace(c,' ')

othelloTxt = othelloTxt.lower()
    
wordCount = {}
oWords = othelloTxt.split(' ')
for c in oWords:
    if c in wordCount.keys():
        wordCount[c] += 1
    else:
        wordCount[c] = 1
```

```python
characters = [
    'OTHELLO',
    'LAGO',
    'DESDEMONA',
    'MICHAEL',
    'DUKE',
    'BRABANTIO',
    'GRATIANO',
    'LODOVICO',
    'RODERIGO',
    'MONTANO',
    'CLOWN',
    'EMILIA',
    'BIANCA'
]
```

```python
import pandas as pd

words = []
counts = []
for key,val in wordCount.items():
    words.append(key)
    counts.append(val) 

datacols = {'word':words,
     'count':counts
    }

df = pd.DataFrame(datacols)
df.sort_values(by='count', ascending = False)

df
```
