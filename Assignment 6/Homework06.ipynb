```python
import requests
res = requests.get('https://www.gutenberg.org/cache/epub/2264/pg2264.txt')
html_page = res.text
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
token_txt = sent_tokenize(html_page)
words = []
for s in token_txt:
    for w in word_tokenize(s):
        words.append(w)
print(words)
```

```python
from nltk.corpus import stopwords
from string import punctuation
nltk.download('stopwords')
myStopWords = list(punctuation) + stopwords.words('english')
wordsNoStop = []
for i in words:
    if i not in myStopWords:
        wordsNoStop.append(i)
```

```python
from collections import Counter
countingWords = Counter(wordsNoStop)
sortedWords = sorted(countingWords.items(),key=lambda x: x[1])
topTwenty=[]
for i in range(len(sortedWords)-1,len(sortedWords)-21,-1):
    topTwenty.append(sortedWords[i][0])
print(topTwenty)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
morestops = ['``',"''","'re","'s","'re",'``',"''","'ll","--","\'\'","...",
             "n\'t",'one','would','use','subject','from',"\'m","\'ve"]
```

```python
myStopWords = list(punctuation) + stopwords.words('english') + morestops
```

```python
movieWordList=[]
for i in reviewList:
    movieWordList.append([w for w in word_tokenize(i.lower()) if w not in myStopWords])
```

```python
from nltk.stem.porter import PorterStemmer
p_stemmer = PorterStemmer()

stemmedWordsList = []
for i in movieWordList:
    stemmedWordsList.append([p_stemmer.stem(w) for w in i])
```

```python
!pip install gensim
from gensim import corpora, models
import gensim
```

```python
dictionary = corpora.Dictionary(stemmedWordsList)
corpus = [dictionary.doc2bow(text) for text in stemmedWordsList]
ldamodel = gensim.models.ldamodel.LdaModel(corpus, 
                                           num_topics=2, 
                                           id2word = dictionary, 
                                           passes=20)
for i in ldamodel.print_topics(num_topics=2, num_words=20):
    print(i)
```

### Comments:
Some of the main words the model chooses to represent comedy include "hilarious", "ridiculous", and "crazy". This seems like the analysis worked, because these words correspond well to the comedy genre. As for drama movies, the model seems to chose words like "love" "side" and "sensitive". These words make sense to be featured as main representations for drama movies. It seems that there is a distinct contrast between the comedy movie words and the drama movie words.
